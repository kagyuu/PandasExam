* Pandas のデータ構造 は、Series と DataFrame
* Series は、1次元
* DataFrame は、複数の Series からなる

```python
import pandas as pd

s1 = pd.Series([1,2,3,4,5])
print(type(s1))
print(s1)
```

```python
df1 = pd.DataFrame({
    'id' : [1,2,3],
    'name' : ['bob','alice','john'],
    'sex' : ['m','f','x']
})
print(type(df1))
print(df1)
```

* DataFrame のカラム名は、df.columns で取得できる
* 1行分のデータは、[name] [[name]] で取得できる
    * [name] で取得すると Series が取得できる
    * [[name]] で取得すると 1行の DataFrame が取得できる

```python
print(df1.columns)

print(type(df1.name))
print(df1.name)

print(type(df1['name']))
print(df1['name'])

print(type(df1[['name']]))
print(df1[['name']])
```

* pd.read_csv() で csv を読み込むことができる
* そのほかに、Excel、Data Base、などから読み込むことができる
* df.head() で先頭５行表示。引数で指定可
* df.tail() で末尾５行表示。引数で指定可

```python
iris = pd.read_csv('https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv')

print(iris.head())
print(iris.tail())
```

* DataFrame の一部を取り出す
* df.loc[row:row,col:col] は、行インデックスと列インデックスで DataFrame の一部を DataFrame として切り出す
* df.iloc[row_num:row_num, col_num:col_num] は、行番号と列番号で DataFrame の一部を DataFrame として切り出す
* df.at[row,col] は、行インデックスと列インデックスで指定した要素を切り出す
* df.at[row_num, col_num] は、行番号と列番号で要素を切り出す

```python
print(iris.loc[15:20,'sepal_length':'petal_width'])
print(iris.iloc[25:30,0:4])
print(iris.at[15,'sepal_width'])
print(iris.iat[15,1])
```

* df.query() で、行の絞り込みができる
* 欠損値の排除などに使えそうだ

```python
iris.query('sepal_length > 6.5')
```

* df.to_csv() で DataFrame を csv に書き出すことができる

```python
iris.query('sepal_length > 6.5').to_csv('filtered_iris.csv', encoding="UTF-8")
```
